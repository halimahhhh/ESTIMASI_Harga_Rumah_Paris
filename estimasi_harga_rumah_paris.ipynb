# MODELING
modeling dari situs Modelling atau pemodelan adalah proses membuat sebuah model matematis atau statistik yang digunakan untuk menganalisis, memprediksi, atau memahami suatu fenomena atau masalah. Pemodelan dilakukan dengan menggunakan teknik dan algoritma dari ilmu statistik, matematika, dan komputer. 
modeling dapat dilakukan untuk membangun model prediksi harga rumah di Paris berdasarkan berbagai variabel seperti
1. cityCode: Kode POS dari lokasi rumah
2. made: Tahun pembuatan rumah
3. floors: Jumlah tingkat rumah
4. squareMeters: Ukuran rumah dalam meter persegi
5. basement: Luas ruangan bawah tanah
6. attic: Luas loteng
7. garage: Luas garasi
8. numPrevOwners: Jumlah pemilik sebelumnya
9. numberOfRooms: Jumlah kamar
10. hasGuestRoom: Apakah rumah memiliki kamar tamu (1 = Ya, 0 = Tidak)
11. hasYard: Apakah rumah termasuk halaman (1 = Ya, 0 = Tidak)
12. hasPool: Apakah rumah termasuk kolam renang (1 = Ya, 0 = Tidak)
13. isNewBuilt: Apakah rumah sudah pernah direnovasi (1 = Ya, 0 = Tidak)
14. hasStormProtector: Apakah rumah memiliki pelindung badai (1 = Ya, 0 = Tidak)
15. hasStorageRoom: Apakah rumah memiliki gudang (1 = Ya, 0 = Tidak)
16. cityPartRange: Kisaran nilai murah sampai mahal dari lokasi rumah (1-10).

Variabel harga adalah variabel target yang ingin diprediksi.

# seleksi fitur

```python
features = ['cityCode', 'made', 'floors', 'squareMeters', 'basement', 'attic', 'garage', 'numPrevOwners', 'numberOfRooms', 'hasGuestRoom', 'hasYard', 'hasPool', 'isNewBuilt', 'hasStormProtector', 'hasStorageRoom', 'cityPartRange']
x = df[features]
y = df['price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# EVALLUATION
Evaluation atau evaluasi adalah proses untuk mengevaluasi performa suatu model atau sistem. Evaluasi dilakukan untuk menilai sejauh mana model atau sistem yang dibangun mampu mencapai tujuan yang ditetapkan, seperti akurasi, presisi, recall, dan metrik lainnya. Evaluasi dapat dilakukan dengan menggunakan data validasi atau data uji yang tidak digunakan dalam proses pembangunan model atau sistem. Hasil evaluasi dapat digunakan untuk mengambil keputusan dan meningkatkan kualitas model atau sistem yang dibangun.

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```
